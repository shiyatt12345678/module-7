```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd
```

```python
nltk.download("vader_lexicon")
```

```python
df=pd.read_csv("customer_reviews.csv")
df.head()
```

```python
x=SentimentIntensityAnalyzer()
```

```python
def senti(text):
    xyz=x.polarity_scores(text)['compound']
    if xyz >= 0.5:
        return "positive",xyz
    elif  xyz <=-0.5: 
        return "Negative",xyz
    else: 
        return "Netural",xyz
```

```python
df['sentiment']=df['ReviewText'].apply(senti)
```

```python
for index,row in df.iterrows():
    print(f"Review:{row['ReviewText']}\nsentiment:{row['sentiment']}\n")
```
