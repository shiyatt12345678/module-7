```python
import re #regular expression 
import pandas as pd  # pandas is a powerful library for data manipulation and analysis.
import nltk # nlp LIBRUARY
from sklearn.feature_extraction.text import TfidfVectorizer #TfidfVectorizer converts a collection of raw text documents to a matrix of TF-IDF (Term Frequency-Inverse Document Frequency) features.
#Use Case: It transforms text data into numerical format, which is required for machine learning algorithms to process.
from sklearn.model_selection import train_test_split #This function is used to **split your dataset** into **training and testing** subsets.
-# **Use Case**: Helps in evaluating your model’s performance on unseen 
from sklearn.naive_bayes import MultinomialNB #**Purpose**: This is a **Naive Bayes classifier** for multinomially distributed data, especially used in **text classification**.
#- **Use Case**: Works well for problems like spam detection, sentiment analysis, etc.
from sklearn.pipeline import Pipeline#Pipeline` allows you to **chain multiple preprocessing and modeling steps** together so you don’t have to manually fit/transform each step.
- #**Use Case**: Simplifies workflows and makes code cleaner and more maintainable.
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
#Breaking it into words (tokenizing)
#Removing unnecessary words (stopwords)
#Reducing words to their base form (lemmatizing)


from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
df=pd.read_csv("spam_ham_dataset.csv")
```

```python
def clean(text):
    text=text.lower() # Step 1: Lowercasing
    text=re.sub(r'[^\w\s]','',text)# Step 2: Remove punctuation
    tokens=word_tokenize(text) Step 3: #Tokenize
    tokens=[WordNetLemmatizer().lemmatize(word) for word in tokens]# Step 5: Remove stopwords
    tokens=[word for word in tokens if word not in stopwords.words('english')] # Step 6: Join tokens back into a string
    return"".join(tokens)
```

```python
df['cleaned']=df['email'].apply(clean)
```

```python
df
```

```python
X_train,X_test,y_train,y_test=train_test_split(df['email'],df['label'],test_size=0.2,random_state=42)
```

```python
pipeline=Pipeline([('tfidf',TfidfVectorizer()),('Classifier',MultinomialNB())])
#'tfidf': TfidfVectorizer()
# Classifier': MultinomialNB()
# Purpose: This is the actual machine learning model, a Naive Bayes classifier.

# Multinomial Naive Bayes is ideal for text classification, especially when using count or frequency features like TF-IDF.
```

```python
pipeline.fit(X_train,y_train)
```

```python
user_email=input("Enter your email:")
clean_email=clean(user_email)
pred=pipeline.predict([clean_email])#This line makes a prediction using your trained pipeline. Here's the full explanation:


print("Result:",pred)
```
